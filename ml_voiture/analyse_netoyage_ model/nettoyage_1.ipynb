### Diviser la colonne 'CarName' en deux  --> Supprimer la colonne 'CarName'  --> RÃ©organiser les colonnes du DataFrame


```python
# Diviser la colonne 'CarName' en deux
df[['Marque', 'Modele']] = df['CarName'].str.split(' ', n=1, expand=True)
# Supprimer la colonne 'CarName'
df = df.drop(['CarName' ,"car_ID"], axis=1)
# RÃ©organiser les colonnes du DataFrame
df = df.reindex(columns=['Marque', 'Modele',	"price"	,	"symboling"	,"fueltype",	"aspiration"	,"doornumber"	,"carbody",	"drivewheel"	,"enginelocation",	"wheelbase",	"carlength",	"carwidth"	,"carheight",	"curbweight"	,"enginetype",	"cylindernumber"	,"enginesize",	"fuelsystem",	"boreratio",	"stroke",	"compressionratio"	,"horsepower",	"peakrpm"	,"citympg",	"highwaympg"])


# Afficher le DataFrame rÃ©sultant
df
```

### Convertir les colonnes de pouces en cm


```python
# Convertir les colonnes de pouces en cm
cols = ['wheelbase', 'carlength', 'carwidth', 'carheight']
df[cols] = df[cols].apply(lambda x: x * 0.0254)
df
```

### Convertir les colonnes de pouces en cm

```python
# Convertir les colonnes de pouces en cm
cols = ["curbweight"]
df[cols] = df[cols].apply(lambda x: x * 0.454)
df
```

### Convertir les colonnes de mpg en L/100km

```python
# Convertir les colonnes de mpg en L/100km
cols = ['citympg', 'highwaympg']
df[cols] = (235.214 / df[cols]).round(2)

# Renommer les colonnes
df = df.rename(columns={'citympg': 'consommation_ville', 'highwaympg': 'consommation_autoroute'})
df
```

### Convertir la colonne en litres

```python
# Convertir la colonne en litres
df['enginesize'] = df['enginesize'] * 0.0163871

# Renommer la colonne
df = df.rename(columns={'enginesize': 'litres'})
```
